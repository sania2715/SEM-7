```python
"""
Implement Gradient Descent Algorithm to find the local minima of a function.
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

"""
```

```python
#Importing Necessary Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the function
def f(x):
    return (x + 3) ** 2
```

```python
# Define the derivative of the function
def df(x):
    return 2 * (x + 3)
```

```python
# Gradient Descent Algorithm
def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    history = [x]  # To keep track of the points visited
    for _ in range(num_iterations):
        x = x - learning_rate * df(x)  # Update the point
        history.append(x)
    return x, history
```

```python
# Parameters
starting_point = 2.0  # Starting point
learning_rate = 0.1   # Learning rate
num_iterations = 20   # Number of iterations
```

```python
# Run Gradient Descent
minima, history = gradient_descent(starting_point, learning_rate, num_iterations)
```

```python
# Output the results
print(f"Local minima found at x = {minima:.4f}")
print(f"Function value at minima: y = {f(minima):.4f}")
```

```python
# Optional: Plotting the function and the descent path
x_values = np.linspace(-6, 1, 100)
y_values = f(x_values)
```

```python
plt.plot(x_values, y_values, label='y = (x + 3)^2', color='blue')
plt.scatter(history, f(np.array(history)), color='red', label='Gradient Descent Path')
plt.title('Gradient Descent on y = (x + 3)^2')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.show()
```
